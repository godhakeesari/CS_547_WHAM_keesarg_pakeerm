```
from IPython.display import HTML
import base64
import io

def show_local_mp4_video(file_name, width=720, height=480):
    video_encoded = base64.b64encode(io.open(file_name, 'rb').read()).decode('ascii')
    return HTML(f'''
    <video width="{width}" height="{height}" controls>
        <source src="data:video/mp4;base64,{video_encoded}" type="video/mp4">
    </video>
    ''')
```

```
# Show input video
# If visualization fails, try reducing the video resolution and quality as:
# !ffmpeg -i /content/WHAM/examples/IMG_9732.mov -vf "scale=iw/2:ih/2" -vcodec libx264 -crf 30 examples/IMG_9732_mini.mov
# Try same with output videos.

mp4 = '/content/WHAM/examples/IMG_9732.mov'
show_local_mp4_video(mp4, width=360, height=640)
```

```
# Show reconstructed results

mp4 = '/content/WHAM/output/demo/IMG_9732/output.mp4'
show_local_mp4_video(mp4, width=720, height=640)
```
