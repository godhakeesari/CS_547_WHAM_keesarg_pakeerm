```
from IPython.display import HTML
import base64
import io

def show_local_mp4_video(file_name, width=720, height=480):
    video_encoded = base64.b64encode(io.open(file_name, 'rb').read()).decode('ascii')
    return HTML(f'''
    <video width="{width}" height="{height}" controls>
        <source src="data:video/mp4;base64,{video_encoded}" type="video/mp4">
    </video>
    ''')
```

```
import re

log_path = "/content/WHAM/eval_3dpw.log"

paper_results = {
    "PA-MPJPE": 35.9,
    "MPJPE": 57.8,
    "PVE": 68.7,
    "ACCEL": 6.6
}

with open(log_path, "r", errors="ignore") as f:
    log_text = f.read()

match = re.search(
    r"PA_MPJPE:\s*([0-9.]+),\s*MPJPE:\s*([0-9.]+),\s*PVE:\s*([0-9.]+),\s*ACCEL:\s*([0-9.]+)",
    log_text
)

if match:
    your_results = {
        "PA-MPJPE": float(match.group(1)),
        "MPJPE": float(match.group(2)),
        "PVE": float(match.group(3)),
        "ACCEL": float(match.group(4)),
    }

    print("=" * 65)
    print("               RESULT - 3DPW")
    print("=" * 65)
    print(f"{'Metric':<15}{'Your Result':<15}{'Paper Result':<15}")
    print("-" * 65)

    for key in your_results:
        unit = "mm" if key != "ACCEL" else "m/s²"
        print(f"{key:<15}{your_results[key]:<15.2f}{paper_results[key]:<15.2f} {unit}")

    print("=" * 65)
else:
    print("Metrics not found in log.")
```
